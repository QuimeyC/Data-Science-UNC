```python
columnas_a_eliminar = ['EmployeeNumber', 'StandardHours', 'Over18'] 
df = df.drop(columnas_a_eliminar, axis=1)
```

```python
fig = plt.figure(figsize=(20,5))
ax = plt.axes()
ax = sns.lineplot(x='Age', y='RelationshipSatisfaction', data=df)
plt.xticks(rotation=90)
plt.title(" Satisfaction / Edad")
plt.show()
#Se puede observar que hay mas valores extremos en la satisfaccion laboral de los empleados mas jovenes y los mas longevos de edad.
```

```python
fig = plt.figure(figsize=(20,5))
ax = plt.axes()
ax = sns.countplot(y='JobSatisfaction', data=df, order=df.RelationshipSatisfaction.value_counts().index)
plt.xticks()
plt.title("Satisfaccion de los empleados")
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))
ax = plt.axes()
ax = sns.boxplot(x='MonthlyIncome',  data=df)
ax.set_title("Distribución de los ingresos mensuales de los empleados")

plt.show()
#El sueldo promedio de los empleados es de $6502.
```

```python

df['attrition_num'] = df['Attrition'].map({'Yes': 1, 'No': 0})
tasa_rotacion = df['attrition_num'].mean()
```

```python

plt.pie(df['attrition_num'].value_counts(), labels=['No', 'Sí'], autopct='%1.1f%%')
plt.title('Distribución de la rotación')
plt.show()
```

```python
sales_department = df[df['Department'] == 'Sales']

tasa_rotacion_sales = sales_department['attrition_num'].mean()
print("Tasa de rotación en el departamento de Ventas:", tasa_rotacion_sales)
# La rotación en el departamento de ventas es mayor que en los demás departamentos.
```

```python

ventas = df[df['Department'] == 'Sales']

conteo_rotacion = ventas['attrition_num'].value_counts()

plt.bar(conteo_rotacion.index, conteo_rotacion.values)
plt.xticks([0, 1], ['No', 'Sí'])
plt.xlabel('Rotación')
plt.ylabel('Número de Empleados')
plt.title('Rotación en el Departamento de Ventas')
plt.show()
# La rotación en el departamento de ventas es mayor que en los demás departamentos.
```

```python
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title('Relación entre satisfacción laboral y rotación')
plt.show()   
#xiste una estrecha relacion entre la rotacion y la satisfaccion laboral, a menor satisfaccion laboral es mayor la rotacion.
```

```python
sns.countplot(x='Education', hue='Attrition', data=df)
plt.title('Tasa de rotación por nivel de educación')
plt.show()
```

```python
df['grupo_edad'] = pd.cut(df['Age'], bins=[0, 30, 40, 50, np.inf], labels=['0-30', '31-40', '41-50', '51+'])

df_agrupado = df.groupby(['Department', 'grupo_edad', 'attrition_num']).size().reset_index(name='conteo')

sns.barplot(x='grupo_edad', y='conteo', hue='attrition_num', data=df_agrupado, hue_order=[0, 1], palette=['green', 'red'])
plt.xlabel('Departamento')
plt.ylabel('Número de empleados')
plt.title('Rotación por departamento y grupo de edad')
plt.legend(title='Rotación', labels=['No', 'Yes'])
plt.show()
#La rotación es mayor para los empleados mas jovenes.
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='YearsWithCurrManager', hue='Attrition', data=df)
plt.title('Tasa de rotación por la cantidad de años con el gerente actual')
plt.show()
#Hay una tasa de rotacion por la cantidad de años con el mismo gerente donde se detaca el numero de años 14(catorce).
```

```python
conteo_por_campo = df['EducationField'].value_counts()

sns.barplot(x=conteo_por_campo.index, y=conteo_por_campo.values)
plt.xlabel('Campo de Estudio')
plt.ylabel('Número de Empleados')
plt.title('Total de Empleados por Campo de Estudio')
plt.xticks(rotation=45) 
plt.show()
# El campo de estudio predominante dentro de los empleados de IBM son las Ciencias de la Vida, seguido por Medicina.
```

```python
df['YearsInCurrentRole_cat'] = pd.cut(df['YearsInCurrentRole'], bins=[0, 5, 10, 15, 20,])

sns.boxplot(x='YearsInCurrentRole_cat', y='JobSatisfaction', data=df)
plt.title('Distribución de la Satisfacción Laboral por Rangos de Años')
plt.xlabel('Rangos de Años en el Rol Actual')
plt.ylabel('Satisfacción Laboral')
plt.show()
```

```python
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df)
plt.title('Distribución de la Distancia al Hogar por Rotación')
plt.xlabel('Rotación')
plt.ylabel('Distancia al Hogar')
plt.show()
#La rotación es mayor para los empleados con mas distancia al trabajo.
```

```python
sns.boxplot(x='PercentSalaryHike', y='JobSatisfaction', data=df)
plt.title('Distribución de la Satisfacción Laboral por Rango de Aumento Salarial')
plt.xlabel('Rango de Aumento Salarial')
plt.ylabel('Satisfacción Laboral')
plt.show()
# El aumento de salario no significa una mejora en la satisfaccion laboral.
```

```python
sns.boxplot(x='OverTime', y='PercentSalaryHike', data=df)
plt.title('Distribución del Porcentaje de Aumento Salarial por Horas Extras')
plt.xlabel('Horas Extras')
plt.ylabel('Porcentaje de Aumento Salarial')
plt.show()
#El conseguir un aumento salarial no conlleva a que el empleado quiera trabajar mas horas ni tampoco menos.
```

```python
sns.boxplot(x='YearsSinceLastPromotion', y='JobSatisfaction', data=df)
plt.title('Distribución de la Satisfacción Laboral por Años desde la Última Promoción')
plt.xlabel('Años desde la Última Promoción')
plt.ylabel('Satisfacción Laboral')
plt.show()
# Los empleados con mayor años sin alguna promocion de puesto de trabajo tienen una satisfaccion menor.
```
